# Subsetting pycistopic

Notebook to subset and keep only cells from D0 and Q1 that are tumour cells

```python
import pycisTopic
import pandas as pd
import os
pycisTopic.__version__
```

```python
os.chdir(os.getcwd())
```

```python
!pwd
```

```python
out_dir = "outs"
os.makedirs(out_dir, exist_ok = True)
```

```python
import pickle
file_path = "../outs/cistopic_obj.pkl" # this is path to the cistopic object with all the cells
with open(file_path, 'rb') as file:
    cistopic_obj = pickle.load(file)

print(cistopic_obj)
```

```python
cistopic_obj.cell_data
```

```python
cells_to_keep = cistopic_obj.cell_data[
    (cistopic_obj.cell_data['cell_type_all'].isin(['Pro-B_cells_D0', 'Pro-B_cells_Q1']) & cistopic_obj.cell_data['sample_id'].isin(['D0', 'Q1']) )
].index.tolist()
```

```python
cistopic_obj.subset(cells=cells_to_keep)
```

```python
print(cistopic_obj)
```

```python
pickle.dump(
    cistopic_obj,
    open(os.path.join(out_dir, "subset_cistopic_obj.pkl"), "wb")
)
```
